```python
import xml.etree.ElementTree as ET
import http.client
```

```python

conn = http.client.HTTPSConnection("www.vrt.be",443)
conn.request("GET", "/vrtnws/nl.sitemap-index.xml")
r1 = conn.getresponse()
#with open("/home/ubuntu/data/tmp/vrt.sitemap","w") as fout:
#    fout.write(r1.read().decode())
sitemap = r1.read().decode("utf-8")
print(sitemap)
```

```python
tree = ET.parse(r1)
```

```python
ns = {
    "vrt": "http://www.sitemaps.org/schemas/sitemap/0.9",
    "vrtnws": "http://www.google.com/schemas/sitemap-news/0.9"
}

for sitemap in tree.findall("vrt:sitemap", ns):
    for loc in sitemap.findall("vrt:loc", ns):
        print(loc.text)
```
